# 1. Introduction

## 1.1 Phase 1 Summary

Phase 1 covered data understanding, preprocessing, and exploratory analysis for the *Estimation of Obesity Levels Based on Eating Habits and Physical Condition* dataset obtained from the UCI Machine Learning Repository. The data was collected from individuals in Mexico, Peru, and Colombia, with approximately 77% of the records synthetically generated using the SMOTE filter to balance the class distribution across all seven obesity categories.

The dataset contained 2,111 observations and 17 variables, consisting of 16 descriptive features and one target variable (`NObeyesdad`). The target variable represents obesity level across seven ordered classes ranging from *Insufficient Weight* to *Obesity Type III*, making this a multi-class classification problem.

Several preprocessing steps were completed to prepare the dataset for predictive modelling. Data types were first examined to ensure all variables were stored appropriately. Missing value analysis was then performed, with no missing values identified across any columns. Duplicate rows were detected and removed to improve overall data quality and reduce redundancy within the dataset.

The dataset was also checked for ID-like variables that could negatively affect machine learning performance. No identifier-style columns were present, as all variables contained meaningful information relevant to obesity classification. Numeric features were further assessed for unrealistic or invalid values, however all values fell within plausible ranges.

Outlier analysis was performed on the eight numeric variables using the Interquartile Range (IQR) method. Detected outliers were replaced with the column median, which reduces the impact of extreme values without heavily affecting the overall distribution of the data. One-hot encoding and standard scaling were intentionally deferred until Phase 2 to avoid data leakage. This ensures that preprocessing transformations are fitted only on the training data after the dataset has been split.

Extensive exploratory analysis and visualisation were also conducted in Phase 1. Twelve visualisations were created across one-variable, two-variable, and three-variable analyses. The one-variable plots examined obesity class distributions, participant age, BMI distributions with WHO cut-off ranges, and per-class weight density curves. Two-variable plots included boxplots of body weight and physical activity frequency against obesity level, a stacked bar chart of snacking behaviour by class, and a heatmap of z-score normalised feature means. Three-variable plots explored relationships between height and weight coloured by obesity class, physical activity frequency split by gender and obesity level, age versus weight with activity frequency represented by point size, and BMI distributions faceted by gender.

The exploratory analysis revealed several noticeable trends within the dataset. BMI, weight, and height showed clear separation across obesity classes, which is expected given how obesity categories are partially defined. Physical activity frequency and snacking behaviour also showed moderate but consistent differences between groups. These findings suggested that the dataset contained enough useful information for supervised classification modelling.

The cleaned dataset was saved as `Phase2_Group08.csv` and is used as the input dataset for this report. Overall, the preprocessing and exploratory analysis completed in Phase 1 established a clean and well-understood dataset that provides the foundation for the feature selection, model fitting, hyperparameter tuning, and model evaluation performed in Phase 2. In this phase, five classification algorithms are applied to the dataset: Decision Tree, Random Forest, K-Nearest Neighbours, Support Vector Machine, and Logistic Regression.

## 1.3 Overview of Methodology

The predictive modelling approach used in this project follows a standard machine learning workflow designed to produce reliable results while avoiding data leakage and allowing a fair comparison between multiple classification models. The overall process consists of data preparation, feature selection, model training, hyperparameter tuning, and model comparison.

### Data Preparation

The cleaned dataset from Phase 1 is first separated into descriptive features (`X`) and the target variable (`y`). The data is then split into training (80%) and test (20%) sets using a stratified random split so that the proportion of each obesity class remains similar across both subsets.

All preprocessing steps are fitted only on the training data before being applied to the test data. This follows the Split → Preprocess → Train → Evaluate approach discussed in class and helps avoid data leakage, where information from the test set could unintentionally influence the training process.

Categorical variables are converted using one-hot encoding through `OneHotEncoder`, with `drop='first'` used to reduce redundancy between encoded variables. Numeric variables are standardised using `StandardScaler` so that features are placed on a similar scale. Standardisation is especially important for models such as K-Nearest Neighbours and Support Vector Machine because these algorithms are sensitive to differences in feature magnitude.

### Feature Selection

Feature selection is performed using `SelectKBest` with the `f_classif` scoring function. This method calculates ANOVA F-scores to measure the relationship between each feature and the target variable.

The F-scores are calculated using the training data only and displayed using a ranked bar chart. A selected number of top-ranked features is then retained and used across all models. Using the same feature set for each algorithm helps keep the comparison between models consistent.

### Model Training and Hyperparameter Tuning

Five supervised classification algorithms are used in this project: Decision Tree (`DecisionTreeClassifier`), Random Forest (`RandomForestClassifier`), K-Nearest Neighbours (`KNeighborsClassifier`), Support Vector Machine (`SVC`), and Logistic Regression (`LogisticRegression`). These algorithms were chosen because they represent different approaches to classification and are commonly used in supervised machine learning problems.

Hyperparameter tuning is performed using `GridSearchCV` with 5-fold stratified cross-validation on the training data. Stratification ensures that each fold maintains a similar class distribution to the original dataset, which is important for a seven-class classification problem.

The tuning metric used is macro-averaged F1-score. This metric gives equal importance to all obesity classes rather than weighting results toward the larger classes in the dataset. The best hyperparameter combination for each model is reported together with tuning plots showing how model performance changes across the parameter grid.

### Model Comparison

After tuning, the final models are evaluated on the held-out test set using accuracy, macro-averaged F1-score, confusion matrices, and classification reports. These measures provide both an overall indication of model performance and a more detailed breakdown of how well each obesity class is classified.

To compare model performance statistically, 10-fold stratified cross-validation is performed for each tuned model to generate a distribution of accuracy scores. Paired t-tests are then applied between model pairs to determine whether the observed performance differences are statistically significant at the α = 0.05 level.

The final results are summarised using tables and boxplots to provide a clearer comparison of performance across all five classification algorithms.
